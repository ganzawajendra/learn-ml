# BAB 1 - Matplotlib Basic
Matplotlib adalah *library* dasar untuk visualisasi data di Python. Di dunia AI/ML, *library* ini sangat krusial untuk melakukan **Exploratory Data Analysis (EDA)**, melihat distribusi data, mendeteksi *outliners*, hingga memantau performa model (*training loss vs validation loss*)

## A. Arsitektur Utama Matplotlib
Dalam Machine Learning, kita wajib menguasai **Object-Oriented (OO) Style** menggunakan fungsi `plt.subplots()`. Gaya ini memisahkan grafik menjadi dua komponen utama:

```python
fig, ax = plt.subplots(figsize=(8, 4))
```

1. **Figure (`fig`):** Kanvas besar atau jendela utama tempat grafik digambar
2. **Axes (`ax`):** Area dalam kanvas tempat data diplot (terdiri dari sumbu X, sumbu Y, label, garis, dll). Satu Figure bisa memiliki banyak Axes 

---

## B. Kustomisasi Plot (Elemen Wajib)
Sebuah plot yang baik haru sbisa dibaca tanpa penjelasan verbal. Berikut adalah fungsi-fungsi wajib untuk memberikan konteks pada grafik:
1. `ax.set_title()` : Memberikan judul utama grafik
2. `ax.set_xlabel()` & ax.set_ylabel() : Memberikan label nama pada sumbu X dan Y
3. `ax.legend()` : Menampilkan legenda penanda data (wajib menyertakan parameter `label` pada saat plotting)
4. `ax.grid()` : Menampilkan garis bantu latar belakang agar mudah dibaca

```python
ax.set_title("Grafik Penduduk")
ax.set_xlabel("Tahun")
ax.set_ylabel("Jumlah Penduduk")
ax.legend(loc="upper right")
```

---

## C. Membuat Multi-Plots
Teknik ini sangat sering digunakan oleh ML Engineer untuk membandingkan dua metrik sekaligus secara berdampingan (misalnya grafik *Loss* vs *Accuracy*)
1. nrows : Jumlah baris plot
2. ncols : Jumlah kolom plot
3. plt.tight_layout() : Fungsi wajib di akhir agar susunan antar plot rapi dan tidak saling bertabrakan

```python
# Membuat 2 plot
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))
```

---

## D. Jenis Gambar Dasar
Matplotlib menyediakan fungsi dasar untuk jenis grafik yang berbeda, tergantung tipe data yang dihadapi:
1. Scatter Plot (`ax.scatter()`) : **Cocok untuk** melihat korelasi antar 2 fitur numerik yang kontinu, atau melihat sebaran hasil clustering

    ```python
    ax.scatter(x_scatter, y_scatter)
    ```

2. Bar Plot (`ax.bar()` atau `ax.barh()`) : **Cocok untuk** membandingkan nilai antar kategori

    ```python
    ax.barh(fitur_penting, skor_kepentingan)
    ```

---

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
# Konsep Pembuatan Plot

x = np.linspace(0, 10, 100)
y = np.sin(x)

# Membuat kanvas (Figure) dan area gambar (Axes)
fig, ax = plt.subplots(figsize=(8, 4)) # figsize menentukan ukuran (lebar, tinggi) dalam inci

# Menggambar grafik di dalam objek ax
ax.plot(x, y, label='Sinus')

plt.show()
````

````python
# Kustomisasi Dasar (Elemen Wajib Ada)

x = np.array([2020, 2021, 2022, 2023, 2024, 2025, 2026])
y = np.array([80, 90, 74, 99, 12, 10, 55])

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, y, label="Grafik Penduduk", color="teal")

# set_title()
# Memberikan judul plot
ax.set_title("Grafik Penduduk")

# set_xlabel()
# Memberikan informasi pada sumbu X
ax.set_xlabel("Tahun")

# set_ylabel()
# Memberikan informasi pada sumbu Y
ax.set_ylabel("Jumlah Penduduk")

# legend()
# Menampilkan keterangnan garis warna -> jika memiliki parameter label
ax.legend(loc="upper right")

# grid()
# Menambahkan garis bantu
ax.grid()

plt.show()
````

````python
# Jenis Plot Dasar

# Scatter -> cocok untuk clustering karena bersifat penyebaran data
# Bar -> cocok untuk membandingkan nilai antar fitur (karena berbentuk bar)
# Plot -> cocok untuk melihat trend data yang kontinu (line)

# Data untuk scatter
x_scatter = np.random.randn(50)
y_scatter = np.random.randn(50)

# Data untuk bar
kelas = ["Kelas A", "Kelas B", "Kelas C"]
rata_nilai = [80, 83, 97]

# Membuat 2 plot
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# Plot Scatter
ax1.scatter(x_scatter, y_scatter, color='purple')
ax1.set_title("Contoh Scatter Plot")
ax1.set_xlabel("Fitur X")
ax1.set_ylabel("Fitur Y")

# Plot Bar
ax2.bar(kelas, rata_nilai)
ax2.set_title("Contoh Bar Plot")
ax2.set_xlabel("Kelas")
ax2.set_ylabel("Nilai Rata-rata")

plt.show()
````
