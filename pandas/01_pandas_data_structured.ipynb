```python
import numpy as np
import pandas as pd
```

```python
# Series -> 1D
# DataFrame -> 2D
# Kolom DataFrame -> kumpulan Series
# Pada ML -> kolom = fitur, baris = sample 
# Data kosong wajib dibersihkan karena ML sangat benci
```

```python
# Series -> dimensi untuk menampung object dengan index
# Series -> 2 komponen utama (data dan index)
# membuat series dari list index otomatis
data_list = [10, 20, 30, 40, 50]
series1 = pd.Series(data_list)
series1
```

```python
# membuat series dari kustom index
hargaBuah = [15000, 25000, 30000]
namaBuah = ["apel", "jeruk", "mangga"]
series2 = pd.Series(data=hargaBuah, index=namaBuah)
series2
```

```python
# membuat series dari dictionary
dataDict = {
    "Jakarta" : 20,
    "Bandung" : 30,
    "Semarang" : 43
}
series3 = pd.Series(dataDict)
series3
```

```python
# membuat series dengan data kosong
data_kosong = [1, 2, 3, np.nan, 4]
series4 = pd.Series(data_kosong) # Otomatis tipe data menjadi float ketika ada NaN
series4
```

```python
# DataFrame -> struktur data berbentuk 2 dimensi
# DataFrame -> ada 3 komponen utama (data, index, column)
# membuat dataframe dari dictionary list
dataPeserta = {
    "Nama" : ["Ganza", "Budi", "Tatang"],
    "Nilai" : [97, 77, 89],
    "Kota" : ["Surakarta", "Bali", "Kudus"]
}
df_peserta = pd.DataFrame(dataPeserta)
df_peserta
```

```python
# membuat dataframe dengan kustom index
id_peserta = ["L200231", "L12931" , "L023904"]
df_kustom = pd.DataFrame(data=dataPeserta, index=id_peserta)
df_kustom
```

```python
# Eksplorasi atribut Dataframe
print("1. Indeks Baris:", df_kustom.index) # Melihat index baris
print("\n2. Daftar Kolom:", df_kustom.columns) # Melihat nama kolom
print("\n3. Dimensi Data (Baris, Kolom):", df_kustom.shape) # Melihat dimensi
print("\n4. Tipe Data per Kolom:\n", df_kustom.dtypes) # Melihat tipe data setiap kolom
print("\n5. Konversi ke NumPy Array:\n", df_kustom.values) # Mengambil data mentah dalam bentuk numpy array
```
