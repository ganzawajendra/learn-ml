## A. Definisi & Konsep Dasar
Di NumPy, kita terbiasa dengan `ndarray` (array multi-dimensi) yang menggunakan indeks bawaan berupa angka ($0, 1, 2, \dots$). Pandas membungkus NumPy array ini menjadi objek yang jauh lebih fleksibel untuk manipulasi data tabular.

Pandas memiliki dua objek fundamental yang menjadi tulang punggung seluruh operasi data:
1. **Series** (1-Dimensi): Seperti satu kolom di Excel atau gabungan NumPy 1D dengan Python Dictionary.
2. **DataFrame** (2-Dimensi): Seperti tabel Excel utuh yang terdiri dari baris dan kolom. Secara teknis, setiap satu kolom di dalam DataFrame adalah sebuah **Series**.

---

## B. Pandas Series (1-Dimension)
Series memiliki dua komponen utama: **Data** (nilainya) dan **Index** (label untuk nilainya). Jika indeks tidak ditentukan, Pandas akan otomatis membuat indeks integer dari $0$ hingga $n-1$.

### Contoh Kode:
```python
import numpy as np
import pandas as pd

# 1. Membuat Series dari List dengan indeks otomatis
data_list = [10, 20, 30, 40]
series_1 = pd.Series(data_list)

# 2. Membuat Series dengan Kustom Indeks (Labeling)
harga_buah = [15000, 25000, 30000]
nama_buah = ["Apel", "Mangga", "Alpukat"]
series_buah = pd.Series(data=harga_buah, index=nama_buah)

# 3. Membuat Series dari Dictionary (Key otomatis jadi Indeks)
data_dict = {"Jakarta": 32, "Bandung": 24, "Surabaya": 34}
series_suhu = pd.Series(data_dict)
```

---

## C. Pandas DataFrame (2-Dimension)
DataFrame adalah struktur data berbentuk tabular yang terdiri dari baris dan kolom. Memiliki 3 komponen utama: **Data**, **Index** (label baris), **Columns** (label kolom).

### Contoh Kode:
```python
# 1. Membuat DataFrame dari Dictionary berisi List
data_peserta = {
    "Nama": ["Sari", "Budi", "Siti"],
    "Nilai_ML": [95, 88, 92],
    "Kota": ["Surakarta", "Jakarta", "Bandung"],
}
df_peserta = pd.DataFrame(data_peserta)

# 2. Membuat DataFrame dengan Kustom Indeks Baris (misal: ID Sampel)
id_peserta = ["ID-001", "ID-002", "ID-003"]
df_kustom = pd.DataFrame(data=data_peserta, index=id_peserta)
```

---

## D. Atribut Penting (Attributes) Objek Pandas
Karena Pandas berbasis OOP, objek Series dan DataFrame memiliki atribut bawaan yang bisa dipanggil **tanpa tanda kurung** `()` untuk memeriksa struktur datanya:
* `df.index` : Melihat label indeks baris
* `df.columns` : Melihat daftar nama kolom (hanya untuk DataFrame)
* `df.shape` : Melihat dimensi data dalam bentuk tuple `(jumlah_baris, jumlah kolom)`
* `df.dtypes` : Melihat tipe data setiap kolom
* `df.values` : Mengambil data mentah dan mengonversinya kembali ke bentuk NumPy array

---

## E. Poin Krusial Hubungan dengan AI/ML
1. Penanganan `NaN` (Missing Value)
Pandas mendeteksi data kosong sebagai `np.nan` (Not a Number). Ketika ada `NaN` dalam suatu kolom integer, Pandas otomatis mengubah tipe data kolom tersebut menjadi `float64`
2. Konsep Immutability & Operasi Matematika
Mayoritas fungsi di Pandas bersifat *immutable* (tidak mengubah data asli, melainkan mengembalikan objek baru), **kecuali jika ditimpa ke variabelnya** `df = df.add(10)`
    * **Operasi Matematika:** `.add()`, `.sub()`, `.mul()`, `.div()` atau simbol matematika biasa (`df + 10`) adalah operasi element-wise (menjumlahkan angka ke setiap sel data), **bukan menambahkan baris baru di bawahnya**.
    * **Menambah Data:** Untuk menggabungkan atau menambah baris baru di akhir, gunakan fungsi `pd.concat([df_awal, df_baru], ignore_index=True)`.
3. Pemisahan Fitur ($X$) dan Target ($y$)
Di industri AI/ML, istilah kolom dan baris berganti nama menjadi:
    * **Columns = Fitur ($X$):** Variabel independen yang digunakan model sebagai petunjuk untuk belajar
    * **Index/Baris = Sample Data:** Penanda unik per data (tidak boleh ikut masuk sebagai Fitur agar model tidak *overfitting*)
    * **Target($y$):** Kolom jawaban yang ingin diprediksi oleh model ML (ditentukan oleh goal dari proyek)

````python
# DataFrame -> struktur data berbentuk 2 dimensi
# DataFrame -> ada 3 komponen utama (data, index, column)
# membuat dataframe dari dictionary list
dataPeserta = {
    "Nama" : ["Ganza", "Budi", "Tatang"],
    "Nilai" : [97, 77, 89],
    "Kota" : ["Surakarta", "Bali", "Kudus"]
}
df_peserta = pd.DataFrame(dataPeserta)
df_peserta
````

````python
# membuat dataframe dengan kustom index
id_peserta = ["L200231", "L12931" , "L023904"]
df_kustom = pd.DataFrame(data=dataPeserta, index=id_peserta)
df_kustom
````

````python
# Eksplorasi atribut Dataframe
print("1. Indeks Baris:", df_kustom.index) # Melihat index baris
print("\n2. Daftar Kolom:", df_kustom.columns) # Melihat nama kolom
print("\n3. Dimensi Data (Baris, Kolom):", df_kustom.shape) # Melihat dimensi
print("\n4. Tipe Data per Kolom:\n", df_kustom.dtypes) # Melihat tipe data setiap kolom
print("\n5. Konversi ke NumPy Array:\n", df_kustom.values) # Mengambil data mentah dalam bentuk numpy array
````
